## 1.Importando Bibliotecas

```python
from bs4 import BeautifulSoup as bs
import requests as rq                 #faz a conexão http com o site
import re                             #importar a Regex
```

## 2.Pegando HTML do Site

```python
html_completo = rq.get("https://webscraping.andreregino.com.br")
html_formatado = bs(html_completo.content)
print(html_formatado.prettify())  #.prettify para printar com identação
```

```python
primeiro_h2 = html_formatado.find("h2").text   #buscar o primeiro <h2> do site
print(primeiro_h2)
```

```python
todas_ocorrencias_h2 = html_formatado.find_all(["h2", "h1"])
for ocorrencia_h2 in todas_ocorrencias_h2:
  print(ocorrencia_h2.text)
```

## 3.Pegando Conteúdo usando Classes e IDs

```python
preco = html_formatado.find_all(class_="card-title pricing-card-title") #usamos "class_" porque a palavra "class" é uma palavra reservada do Python.
print(preco)
print("-----------")
pre_requisito = html_formatado.find(id="pre-requisite-section")
print(pre_requisito)
```

## 4.Pegando conteúdo com filtro de texto

```python
todas_ocorrencias_h2 = html_formatado.find_all(["h2", "h1"], text="Reviews")    #busca do termo exato, ou seja, considerando letras maiúscula e minúsculas.
for ocorrencia_h2 in todas_ocorrencias_h2:
  print(ocorrencia_h2.text)
```

```python
todas_ocorrencias_h2 = html_formatado.find_all(["h2", "h1"], text=re.compile("Para que aprender"))  #utilizando Regex
for ocorrencia_h2 in todas_ocorrencias_h2:
  print(ocorrencia_h2.text)
```

```python
botao = html_formatado.find(class_="price").a['href']   #buscar somente a tag <a> e seu atributo 'href'
print(botao)                                            
```

```python
imagem = html_formatado.find("img")['src']
print(imagem)
```

```python
elemento_da_lista = html_formatado.find("section", class_="content-section").ul.li
print(elemento_da_lista)
```

```python
quarto_elemento_da_lista = html_formatado.select(".content-section ul li:nth-of-type(4)") #função select: busca com CSS Selector Reference.
print(quarto_elemento_da_lista)
```
